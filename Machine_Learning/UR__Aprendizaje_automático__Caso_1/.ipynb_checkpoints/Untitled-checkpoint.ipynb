```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('HousingData1')
```

```python
data
```

```python
data.info()
```

Observe que esta primera vista de los datos nos permite identificar que toodas las variables con las que vamos a trabajar son de tipo numÃ©rico. Tambien podemos obsevar que hay 20 datos los cuales no sabemos si son espacios vacios, o algun NaN

```python
data.describe()
```

Este resumen de datos nos permite ver un poco como es la magnitud de los datos, observe que las medias tienen magnitudes distintas, algunas se encuentran entre 0 y 1, otras entre 10 y 20, y luego hay algunas de 68, 400. Esta diferencia de magnitudes puede ser un problema por lo que podemos empezar a considerar aplicar una escala logaritmica o una regulÃ±arizaciÃ³n para los modelos de regresiÃ³n.

```python
data = data.drop(data.columns[0],  axis = 1)
```

```python
feature = data.columns

# Graficar cada caracterÃ­stica contra la etiqueta
fig, axes = plt.subplots(5, 3, figsize=(15, 20))

for i, ax in enumerate(axes.ravel()):
    if i > 12:
        ax.axis('off')  # Desactivar ejes adicionales
        continue
    ax.scatter(data[feature[i]], data['MEDV'], edgecolor='k', s=20)
    ax.set_ylabel("MEDV")
    ax.set_xlabel(feature[i])
    ax.grid(True)

plt.tight_layout()
plt.show()
```

Con estas grÃ¡ficaspodemos observar el comportamiento de las variables con nuestra variable objetivo, podemos observar que las variables 'CHAS','ZN ,'RAD', tienen comportamientos 'extraÃ±os' en principio pueden ser candidatos a variables que no nos aporten mucho en nuestro modelo. 

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))

# Crear un mapa de calor (heatmap) de la matriz de correlaciÃ³n
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# AÃ±adir tÃ­tulo al mapa de calor
plt.title('Matriz de CorrelaciÃ³n')

# Mostrar el mapa de calor
plt.show()
```

En esta matriz de correlaciÃ³n es claro que la variable menos importante es 'CHAS' por lo que en definitiva vamos a descartarla, tambien vemos una fuerte correlaciÃ³n entre las variables 'TAX' y 'RAD', lo que sigue dando sospecha de que podrÃ­amos quitar 'RAD', digo 'RAD' porque tiene menos correlaciÃ³n con nuestra variable objetivo. Tambien podemos observar que que las variable m'as influyentes den el precio son : ' LSTAP' y 'RM' 

```python
# vamos a aplicarle una escala logarÃ­tmica a los datos

X = np.log1p(data.drop(data.columns[[2, 11]], axis=1))
y = np.log1p(data['MEDV'])
```

```python

# X = data.drop(data.columns[[2, 11]], axis=1)
# y = data['MEDV']
```

```python
X
```

```python
X.describe()
```

```python
# Queremos verificar si los valores que faltan son NaN y en caso de que si quitarlos
(X==np.nan).sum()
```

Al no ser valores NaN vamos a suponer que son valores nulos, al ser solo 20 datos y al aplicar nuestra escala logaritmica nos permite reducir la escala de la  desviaciÃ³n estandar y darnos mÃ¡s seguridad al remplazar esos nulos por la media de los valores

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Vamos a aplicar este mÃ©todo a nuestros datos el cual nos permitirÃ¡ rellenar los espaciÃ³s nulos con la media de la columna 
# correspondiente 
numerical_imputer = SimpleImputer(strategy='mean')
X_train_numerical_imputed = numerical_imputer.fit_transform(X_train)
X_test_numerical_imputed = numerical_imputer.transform(X_test)

# volvemos a convertir los datos en Dataframe para poder correrr nuestro modelo
X_train = pd.DataFrame(X_train_numerical_imputed, columns=X_train.columns)
X_test = pd.DataFrame(X_test_numerical_imputed, columns=X_test.columns)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)


# Calcular Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Calcular Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)


# Calcular el Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular R-squared (RÂ²)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (RÂ²): {r2}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print (f"Coeficientes{model.coef_}")
```

```python
X.columns
```

Para este modelo las variables con mayor valor en maginitud son: 'NOX', 'PTRATIO'

```python
mse_scores = []
r2_scores = []
rmse_scores = []
mae_scores = []
# Iterar a travÃ©s de diferentes valores de alpha
alphas = [0.0001,0.0001,0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alphas:
    # Modelo Ridge
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train, y_train)
    y_pred_ridge = ridge_model.predict(X_test)
    
    # Calcular mÃ©tricas
    mse_scores.append(mean_squared_error(y_test, y_pred_ridge))
    r2_scores.append(r2_score(y_test, y_pred_ridge))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred_ridge)))
    mae_scores.append(mean_absolute_error(y_test, y_pred_ridge))

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(alphas, mse_scores)
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.title('MSE vs Alpha')

plt.subplot(2, 2, 2)
plt.plot(alphas, r2_scores)
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('RÂ²')
plt.title('RÂ² vs Alpha')

plt.subplot(2, 2, 3)
plt.plot(alphas, rmse_scores)
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('RMSE')
plt.title('RMSE vs Alpha')

plt.subplot(2, 2, 4)
plt.plot(alphas, mae_scores)
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MAE')
plt.title('MAE vs Alpha')

plt.tight_layout()
plt.show()
```

Tanto el R^2 como las otras metricas de desempeÃ±o nos indica que entre menor sea el alpha mejor es el rendimiento del modelo 
vamos a quedarnos con alpha = 0.0001

Sin embargo la metrica R^2 no presenta una mejorÃ­a significativa al usar una regularizaciÃ³n ridge

Vamos a repetir el modelo usando validaciÃ³n cruzada para encontrar el mejor alpha y entrenar nuestro modelo que tentrÃ¡ la regularizaciÃ³n Ridge

```python
from sklearn.model_selection import GridSearchCV
# Define una lista de valores de alpha a considerar
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# Crea un modelo Ridge
ridge_model = Ridge()

# Realiza una bÃºsqueda de hiperparÃ¡metros con validaciÃ³n cruzada
grid_search = GridSearchCV(estimator=ridge_model, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

# Imprime el mejor valor de alpha encontrado
best_alpha = grid_search.best_params_['alpha']
print(f"Mejor alpha encontrado: {best_alpha}")

# Entrena el modelo Ridge con el mejor alpha
best_ridge_model = Ridge(alpha=best_alpha)
best_ridge_model.fit(X_train, y_train)

# Realiza predicciones y evalÃºa el modelo en el conjunto de prueba
y_pred = best_ridge_model.predict(X_test)

# Calcular Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Calcular Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)


# Calcular el Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular R-squared (RÂ²)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (RÂ²): {r2}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print (f"Coeficientes{model.coef_}") 
```

Volvemos al mismo caso anterior donde aunque se encontrÃ³ que el mejor alpha es de 0.01, no nos da una mejorÃ­a notable en nestro modelo, sobretodo basandosnos en la metrÃ­ca R^2

```python
lasso_model = Lasso()

# Realiza una bÃºsqueda de hiperparÃ¡metros con validaciÃ³n cruzada
grid_search = GridSearchCV(estimator=lasso_model, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

# Imprime el mejor valor de alpha encontrado
best_alpha = grid_search.best_params_['alpha']
print(f"Mejor alpha encontrado: {best_alpha}")

# Entrena el modelo Lasso con el mejor alpha
best_lasso_model = Lasso(alpha=best_alpha)
best_lasso_model.fit(X_train, y_train)

# Realiza predicciones y evalÃºa el modelo en el conjunto de prueba
y_pred = best_lasso_model.predict(X_test)

# Calcular Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Calcular Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calcular el Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular R-squared (RÂ²)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (RÂ²): {r2}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Coeficientes: {best_lasso_model.coef_}")
```
