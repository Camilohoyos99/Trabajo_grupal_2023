# CreaciÃ³n test suite Viaje a Rumania

```python
import numpy as np
from copy import deepcopy
from utils import ListaPrioritaria
import json
from ambientes import ViajeRumania

%reload_ext autoreload
%autoreload 2
```

```python
inicial = 'Pitesti'
v = ViajeRumania(estado_inicial=inicial, final='Bucharest')
v.render(inicial)
```

```python
ciudad_inicial = 'Pitesti'
objetivos = []
estados = ListaPrioritaria()
estados.push(ciudad_inicial, 0)
explorados = [ciudad_inicial]
puz = ViajeRumania(estado_inicial=ciudad_inicial, final=inicial)
profundidad = 0
while profundidad < 7:
    contador = 0
    if len(estados) == 0:
        break
    inicial, profundidad = estados.pop(with_value=True)
    print(profundidad, '->', inicial)
    print('profundidad ->', profundidad)
#    puz.render(inicial)
    acciones = puz.acciones_aplicables(inicial)
    for accion in acciones:
        objetivos1 = {"env_class":'ViajeRumania'}
        contador += 1
        objetivos1["name"] = f'vrum_{profundidad+1}_{contador}'
        objetivos1["pasos"] = profundidad + 1
        estado = puz.transicion(inicial, accion)
        if estado not in explorados:
            estados.push(estado, profundidad+1)
            explorados.append(estado)
            objetivos1["parameters"] = {
                "estado_inicial":ciudad_inicial,
                "ciudad_objetivo":estado
            }
#            puz.render(estado)
            objetivos.append(objetivos1)
            print(objetivos1)
```

```python
f = open('test_suite_viajerumania.json', 'w')
json.dump(objetivos, f)
f.close()
```

# CreaciÃ³n test suite ReyTorreRey

```python
import numpy as np
from copy import deepcopy
from busqueda import ListaPrioritaria
import json
from juegos import ReyTorreRey
import chess

%reload_ext autoreload
%autoreload 2
```

```python
pl = ' w'
objetivos = {'reqs':['import chess']}
objetivos["games"] = []
for i in range(100):
    objetivo = {}
    objetivo["game_class"] = 'ReyTorreRey'
    objetivo["name"] = f'random_{i}'
    fila = np.random.randint(0,5)
    columna = np.random.randint(0,7)
    tbl = ''
    for j in range(6):
        if j != fila:
            tbl += '8/'
        else:
            inicial = str(columna) if columna > 0 else ''
            final = str(8 - columna - 1) if 7-columna > 0 else ''
            tbl += f'{inicial}k{final}/'
    tbl += '6K1/7R'
    tablero = f'chess.Board("{tbl+pl}")'
    objetivo["parameters"] = {'estado_inicial':tablero}
    objetivos["games"].append(objetivo)
```
